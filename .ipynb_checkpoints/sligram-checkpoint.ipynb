```python
# encontrar o tamanho
N = len(Base_Dados['fase 7040 Hz'])

#frequencia de amostragem
fs = 7040

#vetor tempo
T = 1/fs

t = np.arange(0, N/fs, T)
t
```

```python
# aplicando a fft
f = np.fft.fftfreq(N,T)
transf = np.fft.fft(Base_Dados['fase 7040 Hz'])
transf = np.abs(transf)
transf
```

```python
#plot do dado após a transformada
plt.figure(figsize=(16,4))
plt.plot(f[f>0], transf[f>0]*1/N, color='b')
#plt.plot(f, transf, color='b')
plt.grid(True)
plt.xlabel('Frequência')
plt.ylabel('Amplitude')
```

```python
# Parâmetros do filtro
cutoff =  1000 # Frequência de corte desejada em Hz
```

```python
#filtro passa baixa
def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs  # Frequência de Nyquist
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = filtfilt(b, a, data)
    return y


# Filtrar os dados de fase e quadratura
transf_filtrada = lowpass_filter(transf, cutoff, fs)
# Adicione filtragem para outras variáveis de fase e quadratura aqui

plt.figure(figsize=(12, 6))

# Plotar o sinal original e o sinal filtrado
plt.subplot(2, 1, 1)
plt.plot(f[f>0], transf[f>0]*1/N, label='Original', color='r')
plt.plot(f[f>0], transf_filtrada[f>0]*1/N, label='low pass filter', linewidth=2, color='b')
plt.title('Fase Original e Filtrado')
plt.xlabel('Frequencia')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#filtro passa alta
def butter_highpass(cutoff, fs, order=5):
    nyq = 0.5 * fs  # Frequência de Nyquist
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    return b, a

# Função para aplicar o filtro passa-alta
def highpass_filter(data, cutoff, fs, order=5):
    b, a = butter_highpass(cutoff, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Filtrar os dados de fase e quadratura
transf_filtrada_pass_high = lowpass_filter(transf, cutoff, fs)


plt.figure(figsize=(12, 6))

# Plotar o sinal original e o sinal filtrado
plt.subplot(2, 1, 1)
plt.plot(f[f>0], transf[f>0]*1/N, label='Original', color='r')
plt.plot(f[f>0], transf_filtrada_pass_high[f>0]*1/N, label='high pass filter', linewidth=2, color='b')
plt.title('Fase Original e Filtrado')
plt.xlabel('Frequencia')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Função para criar um filtro passa-banda
def butter_bandpass(lowcut, highcut, fs, order=5):
    b1, a1 = butter_lowpass(highcut, fs, order=order)
    b2, a2 = butter_highpass(lowcut, fs, order=order)
    b = np.convolve(b1, b2)
    a = np.convolve(a1, a2)
    return b, a

# Função para aplicar o filtro passa-banda
def bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Parâmetros do filtro passa-banda
lowcut = 1000   # Frequência de corte inferior em Hz
highcut = 1500  # Frequência de corte superior em Hz

# Filtrar os dados de fase e quadratura
transf_filtrada_passabanda = bandpass_filter(transf, lowcut, highcut, fs)

plt.figure(figsize=(12, 6))

# Plotar o sinal original e o sinal filtrado
plt.subplot(2, 1, 1)
plt.plot(f[f>0], transf[f>0]*1/N, label='Original', color='r')
plt.plot(f[f>0], transf_filtrada_passabanda[f>0]*1/N, label='pass band filter', linewidth=2, color='b')
plt.title('Fase Original e Filtrado')
plt.xlabel('Frequencia')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```
